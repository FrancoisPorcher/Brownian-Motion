# Brownian Motion

This is a very short project about Brownian Motion. I was reading some theory about B.M from Walter Appel and I wanted to visualize brownian motion in 3D and interact with it.

### Libraries

```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

# Experiment

```python
def brownian_motion_3d(i, n=1000):
  """
  This function generates a random walk of size n.
  i is just an integer to give a color to each random walk
  The result is then assembled into a Pandas Dataframe for later plotting
  """


  x = np.random.normal(0,1,n)
  x = np.cumsum(x)
  y = np.random.normal(0,1,n)
  y = np.cumsum(y)
  z = np.random.normal(0,1,n)
  z = np.cumsum(z)

  colors = [i] * n

  df  = pd.DataFrame(data = {'x':x, 'y':y, 'z':z, 'colors':colors})

  return df

```

# 3D Visualisation

## 1 random walk

```python
df= brownian_motion_3d(i = 0, n = 10000)

fig = px.line_3d(df, x="x", y="y", z="z", color='colors')
fig.show()
```

## Several random walks

```python
df = brownian_motion_3d(i = 0)

for i in range(1,6):
  dg = brownian_motion_3d(i =i, n = 1000)
  df = pd.concat([df,dg], axis = 0)

fig = px.line_3d(df, x="x", y="y", z="z", color='colors')
fig.show()
```
